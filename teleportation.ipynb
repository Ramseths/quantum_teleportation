# Teleportation

```python
from qiskit import *
from qiskit.visualization import plot_histogram
%matplotlib inline
```

```python
circuit = QuantumCircuit(3,3)
circuit.x(0)

circuit.barrier()
circuit.h(1)
circuit.cx(1,2)

circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.cx(0, 1)
circuit.h(0)

circuit.barrier()
circuit.draw(output='mpl')
```
